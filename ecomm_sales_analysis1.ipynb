```python
import pandas as pd

# Loading the dataset
file_path = "customer.xlsx"
df = pd.read_excel(file_path)

# Data Cleaning & Renaming 
df = df.rename(columns={
    'InvoiceNo': 'order_id',
    'CustomerID': 'customer_id',
    'InvoiceDate': 'date',
    'Description': 'product',
    'Quantity': 'quantity',
    'UnitPrice': 'price',
    'Country': 'country'
})

# Key Performance Indicators 
total_orders = df['order_id'].nunique()
df['revenue'] = df['quantity'] * df['price']
total_revenue = df['revenue'].sum()

print(f"Total Unique Orders: {total_orders}")
print(f"Gross Revenue: ${total_revenue:,.2f}")

# Product Performance Analysis
# Top 5 Products by Revenue
top_revenue = df.groupby('product')['revenue'].sum().sort_values(ascending=False).head(5)

# Top 3 Products by Order Frequency (Popularity)
top_orders = df.groupby('product')['order_id'].nunique().sort_values(ascending=False).head(3)

print("\nTop 5 Products by Revenue:\n", top_revenue)
print("\nTop 3 Products by Order Volume:\n", top_orders)
```
